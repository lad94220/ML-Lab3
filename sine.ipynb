```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = np.load('sine-data/sine_noisy_scale5_noise0.5.npz', allow_pickle=True)

X = data['X']
y = data['y']

plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Noisy Data', alpha=0.6)
plt.title('Noisy Sine Wave Data')

plt.figure()
plt.plot(X, y, 'o', label='Noisy Data', alpha=0.6)
```
