```python
import torch
import numpy as np
from models import MLP

# ====== C·∫§U H√åNH ======
MODEL_PATH = 'FINAL_DEPLOYMENT_MODEL_protein_GAR_20251213_225421.pth'
# MODEL_PATH = 'FINAL_DEPLOYMENT_MODEL_protein_MAE_20251213_230400.pth'
# MODEL_PATH = 'FINAL_DEPLOYMENT_MODEL_protein_GAR-EXP_20251213_231705.pth'
INPUT_DIM = 9        # f1 -> f9
NUM_TARGETS = 1      # RMSD
HIDDEN_SIZES = (64,128,64,32)  # PH·∫¢I GI·ªêNG l√∫c train
DEVICE = 'cpu'       # ho·∫∑c 'cuda'

# ====== LOAD MODEL ======
model = MLP(
    input_dim=INPUT_DIM,
    hidden_sizes=HIDDEN_SIZES,
    num_classes=NUM_TARGETS
)
model.load_state_dict(torch.load(MODEL_PATH, map_location=DEVICE))
model.to(DEVICE)
model.eval()

import pickle
import torch
import numpy as np

# Load scaler
with open('data/protein_scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)

# Load model
model.load_state_dict(torch.load(MODEL_PATH, map_location='cpu'))
model.eval()

def predict_rmsd(raw_features):
    x = np.array(raw_features).reshape(1, -1)
    x = scaler.transform(x)   # üî• B·∫ÆT BU·ªòC
    x = torch.tensor(x, dtype=torch.float32)

    with torch.no_grad():
        y_pred, _ = model(x)

    return y_pred.item()

```

```python
sample = [8424.58,2368.25,0.28111,67.8325,1210471.6295,109.439,3248.22,70,39.0651]
rmsd = predict_rmsd(sample)
print("Predicted RMSD:", rmsd)
```
