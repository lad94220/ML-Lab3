```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# First Impression

```python
# LOAD DATA
def npz_loader(filepath):
	try:
		npz = np.load(filepath)
		
		X = npz["X"]
		y = npz["y"]

		df = pd.DataFrame(X)
		df["y"] = y.ravel()
		df.columns = ["X","y"]

	except Exception as e:
		print(f"Error loading file: {e}")
		return None

	return df
	
df = npz_loader("sine.npz")
```

```python
# DATA DESCRIPTION
print(f"{'-'*8} * DATA HEAD * {'-'*8}")
print(df.head())

print(f"\n{'-'*6} * DESCRIBE DATA * {'-'*6}")
print("Shape:", df.shape)
print("Missing value:", df.isnull().sum().sum())
print(df.describe())
```

# Data Plotting

```python
# SCATTER PLOT
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["X"], y=df["y"])
plt.title(f"Scatter Plot: Sine dataset")
plt.xlabel(f"X")
plt.ylabel("y")
plt.tight_layout()
plt.show()
```

```python
# VARIABLES DISTRIBUTION
plt.figure(figsize=(12, 6))

# Distribution of X
plt.subplot(1, 2, 1)
sns.histplot(df['X'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of X")
plt.xlabel("X value")
plt.ylabel("Frequency")

# Distribution of y
plt.subplot(1, 2, 2)
sns.histplot(df['y'], bins=30, kde=True, color='orange')
plt.title("Distribution of y")
plt.xlabel("y value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
